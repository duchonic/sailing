```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def displacement(x, hullspeed):
    return (1-1.1**-x) * hullspeed
```

```python
def planing(x, planingspeed, maxspeed):
    return maxspeed/2 + np.arctan(x-planingspeed)  * maxspeed/3
```

```python
def foiling(x, foilspeed, maxspeed):
    return maxspeed/2 + np.arctan(x-foilspeed)  * maxspeed/3
```

```python
x = np.linspace(0, 30, 50)
plt.xlim( (0,30) )
plt.ylim( (0,30) )
plt.plot(x, displacement(x,13), x, planing(x,15,20), x, foiling(x,8,20))
```
