```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def calcApparent(wind, boat):
    return np.linalg.norm(wind+boat)
```

```python
def calcSpeed(vector):
    return int(np.linalg.norm(vector))
```

```python
def dotproduct(v1, v2):
  return sum((a*b) for a, b in zip(v1, v2))
def length(v):
  return math.sqrt(dotproduct(v, v))
def angle(v1, v2):
  return math.acos(dotproduct(v1, v2) / (length(v1) * length(v2)))
def angleDegrees(v1, v2):
    return int(math.degrees(angle(v1,v2)))
```

```python
wind = np.array([0,0,0,15]); print('wind:', calcSpeed(wind),'kn')
gust = np.array([0,0,0,10]); print('gust:', calcSpeed(gust),'kn')
upwind = np.array([0,0,8,8]); print('upwind speed:', calcSpeed(upwind),'kn')
downwind = np.array([0,0,15,-15]); print('downwind speed:', calcSpeed(downwind),'kn')
```

```python
print('apparent up:  ', int(angleDegrees(wind, upwind+wind)),'°', \
      'with gust:', int(angleDegrees(wind, upwind+wind+gust)),'°')
print('apparent down:', int(angleDegrees(wind, downwind+wind)),'°', \
      'with gust:', int(angleDegrees(wind, downwind+wind+gust)),'°')
```

```python
print('apparent up:  ', int(calcApparent(wind, upwind)),'kn', \
      'with gust:', int(calcApparent(wind+gust, upwind)),'kn')
print('apparent down:', int(calcApparent(wind, downwind)),'kn', \
      'with gust:', int(calcApparent(wind+gust, downwind)),'kn')
```

```python
soa = np.array([wind, gust+[0,15,0,0], upwind, upwind+wind, upwind+wind+gust])
X, Y, U, V = zip(*soa)
plt.figure()
ax = plt.gca()
ax.quiver(X, Y, U, V, angles='xy',\
          color=['green','blue','grey','green','blue'], \
          scale_units='xy', scale=1)
ax.set_xlim([-1, 40])
ax.set_ylim([-1, 40])
plt.draw()
plt.show()
```

```python
soa = np.array([wind, gust+[0,15,0,0], upwind, upwind+wind, upwind+wind+gust])
X, Y, U, V = zip(*soa)
ax = plt.gca()
qv = plt.quiver(X, Y, U, V, angles='xy',\
          color=['green','blue','grey','green','blue'], \
          scale_units='xy', scale=1, \
          label=['green','blue','grey','green','blue'])
plt.quiverkey(qv, 0.5, 0.5, 0, 'gust', coordinates='axes',\
              color='r', labelcolor='blue')

ax.set_xlim([-1, 40])
ax.set_ylim([-1, 40])


plt.draw()
plt.show()

print(X, Y)
print(U, V)
```
